```python
from thinkdsp import CosSignal, SinSignal

cos_sig = CosSignal(freq=880, amp=1.0, offset=0)
sin_sig = SinSignal(freq=440, amp=2.0, offset=0)
```

```python
from thinkdsp import decorate

cos_sig.plot()
decorate(xlabel='Time (s)')
```

```python
sin_sig.plot()
decorate(xlabel='Time (s)')
```

```python
mix = sin_sig + cos_sig
mix
```

```python
mix.plot()
decorate(xlabel='Time (s)')
```
